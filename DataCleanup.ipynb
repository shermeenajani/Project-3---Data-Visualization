```python
# Dependencies and Setup
import pandas as pd
import json

# Files to Load 
womensFertility = "data/womensedvfert.csv"
femaleLabor = "data/female-labor.csv"
childrenPerWoman = "data/children-per-woman.csv"
# birthRate = "data/birth-rate.csv"

# Read various fertility stats CSVs
fertilityData = pd.read_csv(womensFertility)
laborData = pd.read_csv(femaleLabor)
childrenPerWomanData = pd.read_csv(childrenPerWoman)
# birthRateData = pd.read_csv(birthRate)

fertilityData = fertilityData.drop('Continent', 1)
laborData = laborData.drop('Continent', 1)
childrenPerWomanData = childrenPerWomanData.drop('Continent', 1)

fertilityData = fertilityData.dropna(subset=['Code', 'Fertility', 'Population'])
fertilityData = fertilityData.reset_index(drop=True)

laborData = laborData.dropna(subset=['Code', 'LaborParticipation', 'Population'])
laborData = laborData.reset_index(drop=True)

childrenPerWomanData = childrenPerWomanData.dropna(subset=['Code', 'Fertility', 'Population'])
childrenPerWomanData = childrenPerWomanData.reset_index(drop=True)


fertEnt = pd.unique(fertilityData['Entity'])
laborEnt = pd.unique(laborData['Entity'])
childEnt = pd.unique(childrenPerWomanData['Entity'])


laborData = laborData.drop('Population', 1)
childrenPerWomanData = childrenPerWomanData.drop('Population', 1)

laborData = laborData.drop('GDP', 1)
childrenPerWomanData = childrenPerWomanData.drop('GDP', 1)

childrenPerWomanData = childrenPerWomanData.drop('Fertility', 1)

fertilityDataComp = pd.merge(fertilityData, laborData, how="inner", on=["Entity", "Code", "Year"])
fertilityDataComp = pd.merge(fertilityDataComp, childrenPerWomanData, how="inner", on=["Entity", "Code", "Year"])



fertilityDataCompMYOS = fertilityDataComp.dropna(subset=['MYOS'])
fertCompEnt = pd.unique(fertilityDataCompMYOS['Entity'])

fertilityDataCompMYOS

boolean_series = fertilityDataComp.Entity.isin(fertCompEnt)
fertilityDataCompFiltered = fertilityDataComp[boolean_series]


fertCompEnt = pd.unique(fertilityDataCompFiltered['Entity'])
print(len(fertCompEnt))
```

```python
country_list = []

for country in fertCompEnt:
    country_list.append(country)

fert_data = {'countries': country_list, 'data':[]}


for country in fertCompEnt:
    countryDF = fertilityDataCompFiltered[fertilityDataCompFiltered["Entity"] == country]

#     country_dict = {'country': country, 'year':[],'fertility':[],'MYOS':[], 'population':[]}
    country_dict = {'country': country, 'year':[],'fertility':[], 'population':[], 'labor':[]}

    for i in range(len(countryDF)):
        country_dict['year'].append(int(countryDF["Year"].iloc[i]))
        country_dict['fertility'].append(float(countryDF["Fertility"].iloc[i]))
#         country_dict['MYOS'].append(float(countryDF["MYOS"].iloc[i]))
        country_dict['population'].append(int(countryDF["Population"].iloc[i]))
        country_dict['labor'].append(float(countryDF["LaborParticipation"].iloc[i]))

    fert_data['data'].append(country_dict)

print(fert_data)


```

```python
json_object = json.dumps(fert_data)
  
```

```python
# Writing to sample.json
with open("fertdata.json", "w") as outfile:
    outfile.write(json_object)
```
